```python
import sympy as sp
import math
```

# Symbols



```python
# Make a symbol
x = sp.Symbol("x")
x
```

```python
# Make a simple expression
2 * x + 5
```

```python
# The variable name and the symbol name do not need to be the same (although it is good practice)
y = sp.Symbol("z")
y
```

# Simplification

```python
# SymPy expressions are automatically simplified
2 * x + x - 5
```

```python
# Sine and cosine are not automatically simplified
sp.sin(x)**2 + sp.cos(x)**2 # Always equals 1
```

```python
# Division and fractions
2 * x / 6
```

```python
# Cancellation in division
2 * x / x
```

```python
# Multiplication is not automatically distributed
expr = x * (x + 2)
expr
```

# Expanding Expressions

```python
# Expand the expression
expr.expand()
sp.expand(expr)
```

```python
# Define multiple symbols at once
s, t = sp.symbols("s, t")
s, t
```

```python
poly = t * (s + 2) * (t - 3)
poly
```

```python
# Expand the curve
poly.expand()
```

# Factoring Expressions

```python
expr = x**2 + 2 * x - 15
expr
```

```python
# Factor the polynomial (Only works for polynomials till the 5th degree)
expr.factor()
sp.factor(expr)
```

```python
# Make as many symbols as follows
x_v = sp.symbols("x0:3")
x_v
```

```python
ex = x_v[0]**2 * x_v[1] * x_v[2] + x_v[0] * x_v[1] * x_v[2]
ex
```

```python
# Factor the epxression
sp.factor(ex)
```
