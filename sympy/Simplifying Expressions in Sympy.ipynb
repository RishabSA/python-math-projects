```python
import sympy as sp
```

```python
# Define the symbols x and y
x, y = sp.symbols("x, y")
x, y
```

# Factor and Expand

```python
# Create a polynomial
poly = x**2 + 5 * x + 6
poly
```

```python
# Factor the polynomial
poly_factored = poly.factor()
poly_factored
```

```python
# Expand the factored polynomial
poly_factored.expand()
```

# Expanding Other Expressions

```python
# Define an exponential function
expr = sp.exp(x + y)
expr
```

```python
# Expand the exponential terms
expr.expand()
```

```python
# Trigonometric expression
trig = (sp.cos(x) + sp.sin(y))**2 - sp.cos(x)**2
trig
```

```python
trig.expand()
```

# Avoiding Certain Simplifications

```python
# Trigonometric and exponential expression
trig_and_exp = sp.exp(x + y) + (sp.cos(x) + sp.sin(y))**2 - sp.cos(x)**2
trig_and_exp
```

```python
# Expand everything
trig_and_exp.expand()
```

```python
# Only exapnd the trigonometric terms
trig_and_exp.expand(power_exp=False)
```

# Cancel

```python
# Define the polynomials
p = x**3 + 10 * x**2 + 31 * x + 30
q = x**2 + 12 * x + 35

fraction = p / q
fraction
```

```python
# Factor and cancel out the common factors from the fraction of polynomials
fraction.cancel()
```

# Apart

```python
# Define the polynomials
p = x**3 + 10 * x**2 + 31 * x + 30
q = x**2 + 12 * x + 35

fraction = p / q
fraction
```

```python
# Perform partial fraction decomposition on the rational function
fraction.apart()
```

# General Simplify

```python
# Define an expression
expr = sp.exp(x + y) + (sp.cos(x) + sp.sin(x))**2 - sp.cos(x) - sp.exp(x)
expr
```

```python
# General simplification
expr.simplify()
```
