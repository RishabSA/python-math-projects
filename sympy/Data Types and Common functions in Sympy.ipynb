```python
import sympy as sp
```

```python
# Define a symbol
x = sp.Symbol("x")
x
```

```python
expr = x**(1/3)
expr
```

```python
# This should be exactly x^(1/15)
expr ** (1/5)
```

# Integer Type

```python
# Integer type
expr = x ** (sp.Integer(1) / sp.Integer(3))
expr
```

```python
# This should be exactly x^(1/15)
expr ** (sp.Integer(1) / sp.Integer(5))
```

```python
# The sympy object is used instead of the python object in this scenario
expr ** (sp.Integer(1) / 5)
```

# Rational Type

```python
# Rational object
rational = sp.Rational(1, 3)
rational
```

```python
expr = (x ** sp.Rational(1, 3)) ** sp.Rational(1, 5)
expr
```

# Common Functions

```python
# Exponential function
x = sp.Symbol("x")
sp.exp(x)
```

```python
# Euler's identity
sp.exp(sp.pi * sp.I) + 1 # = 0
```

```python
# Natural logarithmic function
x = sp.Symbol("x")
sp.log(x)
```

```python
sp.exp(sp.log(x))
```

```python
# Trigonometric functions
sp.sin(x) + sp.cos(x) + sp.tan(x) + sp.csc(x) + sp.sec(x) + sp.cot(x)
```

```python
expr = x ** 2 + sp.sin(sp.pi * x) + sp.exp(sp.cos(x))
expr
```
